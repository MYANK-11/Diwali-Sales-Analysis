```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("/content/drive/MyDrive/Data/Diwali Sales Data.csv",encoding='unicode_escape')  # here I have put a parameter i.e. encoding='unicode_escape'  because this csv file I have copy from desktop to drive so basic this parameter will is used to avoid error while opening file

# df.shape  # it will show us the number of rows and columns

# df.head()

# df.info()

df.drop(["Status","unnamed1"],axis=1,inplace=True)  #Steps 1:- To delete unwanted rows. Here Inplace parameter is used to save the changes the whatever we are deleting basically saving the changes

pd.isnull(df).sum() # Steps 2:- To check whether there is any null value in the data or not. Here 'pd.isnull(df).sum()' means 'isnull()' will check whether there is any null value or not , and sum() will sum the null values if any

df.shape

df.dropna(inplace=True)

# pd.isnull(df).sum()  # to cross-check whether all the null values are remove or not


df['Amount']=df['Amount'].astype('int') # step3:- to check whether the datatypes are correct or not. Here in this case amount datatype are was float so we will change the datatype by using this :- df['Amount']=df['Amount'].astype('int')

# df.info()

# df.columns # it will list all the columns that are present in the dataframe

# df.describe()

# df[["Age","Order","Amount"]].describe()  # it will describe the particular columns only


# EDA:-


ax=sns.countplot(x='Gender',data=df) # it will  create a bar graph. here x is not a normal variable it is a parameter used to that specifies which column of the DataFrame df to use for the x-axis in the plot.
for value in ax.containers:
  ax.bar_label(value)

sale_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False) # so here we can divide into three part 1. grouping the gender columns. 2. sum of amount 3. sort amount descending
sns.barplot(x='Gender',y='Amount',data=sale_gen)
```

From the above graph we can see that most of the buyers are females and even the purchasing power of females are greater than men

```python

# Occupation:-

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("/content/drive/MyDrive/Data/Diwali Sales Data.csv",encoding='unicode_escape')

sns.set(rc={'figure.figsize':(7,5)})

ax=sns.countplot(data=df,x='Occupation')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(21,5)})
sns.barplot(data=sales_state,x='Occupation',y='Amount')
```

From the above graph we can see that most of the buyers are working in IT, Healthcare And Aviation Sector

```python

# Product Category :-
sns.set(rc={'figure.figsize':(29,5)})

ax=sns.countplot(data=df,x="Product_Category")

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(29,5)})
sns.barplot(data=sales_state,x='Product_Category',y='Amount')
```

from the above graph we can see that most of the sold product are from food ,clothing and electronics category

```python
sales_state=df.groupby(['Product_ID'], as_index=False) ['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(data=sales_state,x='Product_ID',y='Orders')
```
